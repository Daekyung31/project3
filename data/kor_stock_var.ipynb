```python
#코로나
import pandas as pd

corona = pd.read_excel('코로나바이러스감염증-19_확진환자_발생현황_211211.xlsx')

corona.columns = corona.iloc[3]
corona = corona.drop([0,1,2,3,4])

corona['코로나전일대비']=corona['계(명)']-corona['계(명)'].shift()

corona['일자']=pd.to_datetime(corona['일자'])
corona = corona[corona['일자']>'2020-11-19']

corona['계(명)'] = pd.to_numeric(corona['계(명)'])
corona['코로나전일대비'] = pd.to_numeric(corona['코로나전일대비'])

corona = corona.reset_index(drop=True)

#corona.dtypes
corona
```

```python
#환율
exchange = pd.read_excel('환율변동조회_20211211.xls')
exchange.columns = exchange.iloc[2]
exchange = exchange.drop([0,1,2])

exchange = exchange.reset_index(drop=True)
exchange = exchange[['일자','매매기준율']]
exchange['환율전일대비'] = exchange['매매기준율']-exchange['매매기준율'].shift(-1)

exchange['일자']=pd.to_datetime(exchange['일자'])
exchange['매매기준율'] = pd.to_numeric(exchange['매매기준율'])
exchange['환율전일대비'] = pd.to_numeric(exchange['환율전일대비'])

#exchange.dtypes
exchange
```

```python
#유가
oil = pd.read_csv('일일국제원유가격(20211210).csv', encoding='cp949')

oil.columns=['일자','유종','WTI']
oil = oil.iloc[:-5]

oil['유가전일대비']=oil['WTI']-oil['WTI'].shift()


oil['일자']=pd.to_datetime(oil['일자'])
#oil.dtypes
oil= oil[['일자','WTI','유가전일대비']]
oil
```

```python
#미국 s&p500
us = pd.read_csv('S&P 500 내역.csv',encoding='cp949',thousands = ',')

us['일자']=pd.to_datetime(us['일자'])

#us.dtypes
us
```

```python
merge_co_ex = pd.merge(corona,exchange, how='outer',on='일자')
merge_oil = pd.merge(merge_co_ex, oil, how='outer', on='일자')
merge_us = pd.merge(merge_oil, us, how='outer', on='일자')

kor_var = merge_us[['일자','계(명)','코로나전일대비','매매기준율','환율전일대비','WTI','유가전일대비','종가','등락률']]
kor_var.columns=[['일자','코로나 일일확진자(명)','코로나전일대비(명)','환율(원)','환율전일대비(원)','유가,WTI(달러)','유가전일대비(달러)','미국,S&P 500','미국 등락률(%)']]

kor_var.to_csv('kor_var.csv')
kor_var
```
